````python
from google.colab import auth
from google.colab import userdata
import vertexai
from vertexai.generative_models import GenerativeModel, ChatSession
import time

project_id = userdata.get('gemini_project')
location = "us-central1"

# Authenticate to Google Cloud
auth.authenticate_user(project_id=project_id)

# Employ Gemini Ultra for Inference
vertexai.init(project=project_id, location=location)
model = GenerativeModel("gemini-1.0-ultra")
chat = model.start_chat()
def get_chat_response(chat: ChatSession, prompt: str):
    response = chat.send_message(prompt)
    return response.text

# Perform Inference
prompt = "Write a self-contained python function that finds a solution to the 8 queens chess puzzle."
response = get_chat_response(chat, prompt)

# Block Code from Response
start_index = response.find("```") + 9
end_index = response.rfind("```")
code = response[start_index:end_index]
````

````python
# Execute Code
tic = time.time()
exec(code + "\n" + "print(solve_8_queens())")
toc = time.time()
print(f"Execution took {toc-tic:.8f} seconds")
````
